```python

path_to_results_kmeans= "/N/project/laidel_el_mcv/LAIDEL-MRI-tumor-id/results/kmeans_classified/"
path_to_results_slic = "/N/project/laidel_el_mcv/LAIDEL-MRI-tumor-id/results/slic_classified/"


path_to_results_folder = "/N/project/laidel_el_mcv/LAIDEL-MRI-tumor-id/results/consensus_kmeans_slic_classified/"
```

```python
i=0
for path in pathlib.Path(path_to_results_slic).iterdir():
#only segment clean images
    if path.is_file() :
            print(i)
            i=i+1
            #extract ID of data file, used for naming
            data_id = path.parts[len(path.parts)-1]
            data_id =data_id[0:-4]
            print(data_id)
            #load all classifications of same image

            data_slic= np.load(path)
            data_kmeans = np.load(path_to_results_kmeans+data_id+"_classified.npy")#parsing depends on how each file is named
            #check if file blank
            max_slic = np.max(data_slic)
            print("max slic is ", max_slic)
            max_kmeans = np.max(data_kmeans)
            print("max kmeans is ", max_kmeans)

            #normalize slic and kmeans results
            print("writing")
            data_slic = data_slic / np.max(data_slic)
            data_kmeans = data_kmeans / np.max(data_kmeans)

            #aggregate so averaging can be performed
            combined_data_arr = np.array([data_kmeans, data_slic])#4D array to be average on 1st axis
            consensus_brain= np.mean(combined_data_arr, axis=0)
            print(np.unique(consensus_brain))


            #write data output
            np.save(path_to_results_folder+data_id+".npy", consensus_brain, allow_pickle=True, fix_imports=True)
            #mean_var_list.append(calc_mean_var(flair_2D, flair_kmeans, clusters))
            #data_order_ref.append(data_id)

```
